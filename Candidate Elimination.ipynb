```python
import numpy as np 
import pandas as pd

d = pd.read_csv('car.csv')
attr = np.array(d.iloc[:,0:-1])
print("\nInstances are:\n",attr)
target = np.array(d.iloc[:,-1])
print("\nTarget Values are: ",target)

def learn(attr, target): 
    #specific_h = attr[0].copy()
    specific_h = [0 for i in range(len(attr[0]))]
    #sh=attr[0].copy()
    print("\nInitialization of specific_h and general_h")
    print("\nS[0]: ", specific_h)
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print("\nVersion Space: ",general_h)  

    for i, h in enumerate(attr):
        print("\nInstance", i+1 , "is ", h)
        if target[i] == "yes":
            print("Instance is Positive ")
            if all(specific_h)==0:
                specific_h=attr[i-1].copy()
            for x in range(len(specific_h)): 
                if h[x]!= specific_h[x]:                    
                    specific_h[x] ='?'                     
                    general_h[x][x] ='?'
                   
        if target[i] == "no":            
            print("Instance is Negative ")
            if all(specific_h)==0:
                for x in range(len(specific_h)):
                    general_h[x][x]=0
            else:
                for x in range(len(specific_h)): 
                    if h[x]!= specific_h[x]:                    
                        general_h[x][x] = specific_h[x]                
                    else:                    
                        general_h[x][x] = '?'        
        
        print("S[", i+1, "]", specific_h)         
        print("G[", i+1, "]", general_h)
        print("\n")

    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?']]    
    for i in indices:   
        general_h.remove(['?', '?', '?', '?', '?']) 
    return specific_h, general_h 

s_final, g_final = learn(attr, target)

print("Final Specific_h: ", s_final, sep="\n")
print("Final General_h: ", g_final, sep="\n")
```
